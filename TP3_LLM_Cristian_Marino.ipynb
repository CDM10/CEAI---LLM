```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
from IPython.display import display

def generar_imagen(descripcion_textual):
    # Cargar el modelo preentrenado de Stable Diffusion
    #modelo = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
    modelo = StableDiffusionPipeline.from_pretrained("prompthero/openjourney", torch_dtype=torch.float16)

    modelo = modelo.to("cuda")  # Mueve el modelo a la GPU para acelerar el procesamiento

    # Generar la imagen
    with torch.autocast("cuda"):
        imagen = modelo(descripcion_textual).images[0]
    display(imagen)
    # Guardar la imagen generada
    imagen.save("imagen_generada.png")
    print("Imagen generada y guardada como 'imagen_generada.png'")

# Ejemplo de uso
#descripcion = "Un gato montado en una bicicleta en un día soleado"
#descripcion = "A cat riding a bike on a sunny day"
#descripcion = "A playful, fluffy cat confidently riding a vintage bicycle through a vibrant park on a sunny day, surrounded by blooming flowers and tall green trees, with a clear blue sky and a warm, cheerful atmosphere."
descripcion = "Futuristic spacecraft inspired by the design of the Starship Enterprise, flying majestically through the vastness of space, with intricate metallic details, glowing blue and white lights, and a backdrop of distant stars and nebulae."
#descripcion = "uss enterprise in space"
generar_imagen(descripcion)
```
